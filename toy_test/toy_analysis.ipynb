```python
"""
Toy Data set to test OC-SVM and KECA Techniques


https://scikit-learn.org/stable/auto_examples/miscellaneous/plot_anomaly_comparison.html#sphx-glr-auto-examples-miscellaneous-plot-anomaly-comparison-py

OC-SVM is known to be sensitive to outliers. It may not perform very well for outlier detection. It's best suited for novelty detection where training data is not contaminated by outliers. With the right hyperparametres OC-SVM might give some useful results. 

15% of samples generated from random uniform noise

"""
```

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs, make_circles
```

```python
n_samples = 300
outliers_fraction = 0.15
n_outliers = int(outliers_fraction * n_samples)
n_inliers = n_samples - n_outliers

svm_obj = svm.OneClassSVM(
    nu=outliers_fraction,
    kernel="rbf",
    gamma=0.1)
```

```python
x, y = make_circles(
    n_samples=n_samples, 
    noise=outliers_fraction, 
    factor=0.3, 
    random_state=571)

fig1, ax1 = plt.subplots()
ax1.set_title("Original Data")
ax1.plot(x[y == 0, 0], x[y == 0, 1], "ro")
ax1.plot(x[y == 1, 0], x[y == 1, 1], "bo")
ax1.set_xlabel("$x_1$")
ax1.set_ylabel("$x_2$")

```

```python
svm_obj.fit(x)
y_pred = svm_obj.predict(x)
```

```python
fig2, ax2 = plt.subplots()
ax2.set_title("SVM Prediction and Original Data")
ax2.plot(x[y_pred == 1, 0], x[y_pred == 1, 1], "go")
ax2.plot(x[y_pred == -1, 0], x[y_pred == -1, 1], "yo")
ax2.set_xlabel("$x_1$")
ax2.set_ylabel("$x_2$")
```
